```python
import numpy as np
import matplotlib.pyplot as plt

def map_value(x, in_min, in_max, out_min, out_max):
    """Mapeia o valor x da faixa [in_min, in_max] para [out_min, out_max]."""
    return (x - in_min) * (out_max - out_min) / (in_max - in_min) + out_min

# Exemplo de uso
in_min, in_max = 0, 1023
out_min, out_max = 1000, 5000

# Vetor de entrada
inputs = np.linspace(in_min, in_max, 100)
outputs = [map_value(x, in_min, in_max, out_min, out_max) for x in inputs]

# Mostrar alguns exemplos
for val in [0, 512, 1023]:
    print(f"Entrada {val} → Saída {map_value(val, in_min, in_max, out_min, out_max):.2f}")

# Plot para visualização
plt.figure(figsize=(8,4))
plt.plot(inputs, outputs, label="Função de mapeamento")
plt.scatter([0, 1023], [map_value(0, in_min, in_max, out_min, out_max),
                        map_value(1023, in_min, in_max, out_min, out_max)],
            color='r', zorder=5, label="Limites")
plt.xlabel("Entrada")
plt.ylabel("Saída mapeada")
plt.title("Função equivalente a map() do Arduino")
plt.grid(True)
plt.legend()
plt.show()
```
